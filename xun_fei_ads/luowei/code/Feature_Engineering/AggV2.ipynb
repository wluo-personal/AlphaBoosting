```python
__file__=''
import os,sys
import pandas as pd
sys.path.append(os.path.join(os.path.dirname(__file__),'../LIB/'))
from env import FILE
```

```python
train = pd.read_pickle(FILE.train_ori.value)
print('train shape is: {}'.format(train.shape))
test = pd.read_pickle(FILE.test_ori.value)
print('test shape is: {}'.format(test.shape))

X = pd.concat([train.drop(['click'],axis=1),test])
print(X.shape)
```

```python
def print_des(df,col):
    try:
        pmax = df[col].max()
        pmin = df[col].min()
        punique = df[col].nunique()
        punique_rate = len(df) / punique
        print('max is :{}'.format(pmax))
        print('min is: {}'.format(pmin))
        print('num of unique value is: {}'.format(punique))
        print('rate of unique value is: {}'.format(punique_rate))
    except:
        print('error in column {}'.format(col))
```

# View Basic Info

```python
for col in X.columns:
    print('######################')
    print('train:-----------{}'.format(col))
    print_des(train,col)
    print('test:-----------{}'.format(col))
    print_des(test,col)
    print('all:-----------{}'.format(col))
    print_des(X,col)
    print('@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@')
```

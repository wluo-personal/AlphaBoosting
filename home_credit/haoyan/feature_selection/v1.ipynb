# corr

```python
def _apply_group(args):
    func, x, y = args
    return f(x, y)

def apply_by_multiprocessing(func, **kwargs):
    workers = kwargs.pop('workers')
    col = kwargs.pop('col')
    pool = multiprocessing.Pool(processes=workers)
    a = []
    for i in range(len(col)):
        for j in range(i+1, len(col)):
            a.append((func, col[i], col[j]))
    result = pool.map(_apply_group, a)
    pool.close()
    return result

def f(x, y, **kwargs):
    return train[x].corr(train[y]), x, y

x = apply_by_multiprocessing(f, workers=4, col=train.columns.values[:10])
```

```python
x.to_pickle(PATH + 'feature_selection/corr.pkl')
```

```python
inst = inst.groupby(['SK_ID_PREV', 'NUM_INSTALMENT_NUMBER', 'NUM_INSTALMENT_VERSION']).agg({'AMT_INSTALMENT': 'max', 'DAYS_ENTRY_PAYMENT': 'max', 'AMT_PAYMENT': 'sum', 'DAYS_INSTALMENT': 'max', 'SK_ID_CURR': 'max'}).reset_index()

inst = inst.groupby(['SK_ID_PREV', 'NUM_INSTALMENT_NUMBER', 'DAYS_INSTALMENT']).agg({'AMT_PAYMENT': 'max', 'AMT_INSTALMENT': 'sum', 'DAYS_ENTRY_PAYMENT': 'max', 'SK_ID_CURR': 'max'}).reset_index()

inst.dtypes
```

```python
# early is <0, on time =0, late >0
inst['LATE_DAYS'] = inst['DAYS_ENTRY_PAYMENT'] - inst['DAYS_INSTALMENT']
inst['LATE_DAYS_SIGN'] = (np.sign(inst['LATE_DAYS']) > 0) * inst['LATE_DAYS']
inst.dtypes
```

```python
payment_series = inst.groupby(['SK_ID_PREV', 'NUM_INSTALMENT_NUMBER', 'LATE_DAYS_SIGN']).agg({'AMT_PAYMENT': 'sum', 'DAYS_ENTRY_PAYMENT': 'max', 'AMT_INSTALMENT': 'max', 'DAYS_INSTALMENT': 'max', 'SK_ID_CURR': 'max'}).reset_index()
payment_series['AMT_CUM_PAYMENT'] = payment_series.groupby(['SK_ID_PREV', 'NUM_INSTALMENT_NUMBER'])[['AMT_PAYMENT']].cumsum()
payment_series['AMT_LATE_PAYMENT'] = payment_series['AMT_INSTALMENT'] - payment_series['AMT_CUM_PAYMENT']
payment_series['RATE_LATE_PAYMENT'] = payment_series['AMT_LATE_PAYMENT'] / (payment_series['AMT_CUM_PAYMENT']+1)
payment_series['DAYS_ENTRY_PAYMENT'] = payment_series[['DAYS_ENTRY_PAYMENT', 'DAYS_INSTALMENT']].max(axis=1)
payment_series.sort_values('DAYS_ENTRY_PAYMENT', ascending=False)
payment_series.dtypes
```

```python
import pandas as pd
import numpy as np
```

```python
array1 = np.random.uniform(size=40000)
array2 = np.random.uniform(size=40000)
```

```python
df = pd.DataFrame({'array1':array1,'array2':array2})
```

```python
def apply_rank(df,col):
    rank = df[col].rank()
    return df[col] * (rank + 1) / len(df)
```

```python
df['array1_rank'] = apply_rank(df,'array1')
df['array2_rank'] = apply_rank(df,'array2')
```

```python
df['array1_array_2_55'] = (df['array1'] * 5 + df['array2'] * 5) / 10
df['array1rank_array_2rank_55'] = (df['array1_rank'] * 5 + df['array2_rank'] * 5) / 10
df['array1_array_2_28'] = (df['array1'] * 2 + df['array2'] * 8) / 10
df['array1rank_array_2rank_28'] = (df['array1_rank'] * 2 + df['array2_rank'] * 8) / 10
```

```python
df['array1_array_2_55_RANK'] = df['array1_array_2_55'].rank()
df['array1rank_array_2rank_55_RANK'] = df['array1rank_array_2rank_55'].rank()
df['array1_array_2_28_RANK'] = df['array1_array_2_28'].rank()
df['array1rank_array_2rank_28_RANK'] = df['array1rank_array_2rank_28'].rank()
```

## Save By MyRatio

```python
threds = 35
col = 'meByCatRatio'
df_selected = df[df[col] > threds].copy()
print(df_selected.shape)
selected_col = list(df_selected['colName'].values)
prefix = col + '_' + str(threds)+'colNUM_'+str(len(selected_col))
```

```python
train_save_name = '../../data/fill_na_410/train_{}.pkl'.format(prefix)
test_save_name = '../../data/fill_na_410/test_{}.pkl'.format(prefix)
train_save = train_ori.copy()
test_save = test_ori.copy()
for col in tqdm(selected_col):
    train_save[col] = train_na[col].copy()
    test_save[col] = test_na[col].copy()
train_save.to_pickle(train_save_name)
print(train_save_name)
print(train_save.shape)
test_save.to_pickle(test_save_name)
print(test_save_name)
print(test_save.shape)
```
